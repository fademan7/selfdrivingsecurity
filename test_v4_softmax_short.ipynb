```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils import to_categorical
##from keras.datasets import mnist
```

```python
import numpy as np
```

```python
x1 = np.loadtxt('./X_train_short.txt')
x2 = np.loadtxt('./X_test_short.txt')
y_train = np.loadtxt('./y_train_short.txt')
y_test = np.loadtxt('./y_test_short.txt')
## user_id, pc_throttle, pc_brake, pc_steering, pc_speed, pc_lap_distance
X_train = x1.reshape(9615, 3)
X_test = x2.reshape(5685, 3)
print(str(X_train.shape))
print(str(X_test.shape))
```

가속값, 속도, 위치 : 3가지로 구성된 배열로 학습 데이터는 9615, 테스트 데이터는 5685개의 데이터를 사용

```python
print("train data (count, row, column) : " + str(X_train.shape) )
print("test data  (count, row, column) : " + str(X_test.shape) )
```

학습 데이터의 하나를 샘플로 확인

```python
input_dim = 3 
X_train = X_train.reshape(9615, input_dim) 
X_test = X_test.reshape(5685, input_dim)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

학습 시, y값과의 cross entropy를 측정해야하므로, 아래의 코드를 실행하여 y를 one hot encoding으로 변환

```python
num_classes = 11
y_train = to_categorical(y_train, num_classes)
y_test = to_categorical(y_test, num_classes)
```

one hot encoding으로 변환되어 클래스 갯수만큼의 길이를 갖는 벡터로 변경이 되었고 해당되는 인덱스의 값 확인

```python
print(y_train[0])
print(y_test[0])
```

케라스의 Sequential()을 사용하여 소프트맥스를 구현

```python
model = Sequential() 
model.add(Dense(input_dim=input_dim, units = 11, activation='softmax'))
```

categorical_crossentropy를 비용함수로 사용한 경사하강법으로 최적의 W와 biases를 학습

```python
model.compile(optimizer='sgd', loss='categorical_crossentropy', metrics=['accuracy']) 
model.fit(X_train, y_train, batch_size=2048, epochs=100, verbose=0)
```

테스트를 진행하여, 정확도를 측정

```python
score = model.evaluate(X_test, y_test) 
print('Test accuracy:', score[1])
```
